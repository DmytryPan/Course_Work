```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from tensorflow import keras
```

Цифры MNIST

```python
from keras.datasets import mnist
(X_train, Y_train), (X_test, Y_test) = mnist.load_data()
```

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
X_train[0]
```

```python
Y_train[0]
```

```python
fig, ax = plt.subplots(1,10, figsize=(15,8))

for i in range(10):
  img = X_train[Y_train==i][0]
  ax[i].imshow(img)
  ax[i].set_title(f"Цифра {i}")
```

```python

Y_train = keras.utils.to_categorical(Y_train, 10)
Y_test = keras.utils.to_categorical(Y_test, 10)

print(Y_train[0])
print(Y_test[0])
```

```python
X_train = X_train / 255
X_test = X_test /255

X_train = X_train.reshape(X_train.shape[0],1,28,28)
X_test = X_test.reshape(X_test.shape[0],1,28,28)

print(X_train.shape)
print(X_test.shape)
```

```python
from keras.layers import Dense, Conv2D, MaxPooling2D,Flatten, Dropout, InputLayer
from keras import Sequential
model = Sequential([
    InputLayer(input_shape = (1,28,28)),
    Conv2D(32,(2,2),padding='same', activation='relu'),
    MaxPooling2D(padding='same'),
    Conv2D(64,(2,2),padding='same', activation='relu'),
    MaxPooling2D(padding='same'),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.25),
    Dense(10, input_dim=300, activation='softmax')
    ]
)
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(X_train,
                    Y_train,
                    epochs=20,
                    batch_size=256,
                    validation_split=0.2,
                    verbose=1,)
```

```python
train_loss, train_acc = model.evaluate(X_train, Y_train)
test_loss, test_acc = model.evaluate(X_test, Y_test)

print(f"Потери на обучающей выборке: {train_loss}, Аккуратность: {train_acc}")
print(f"Потери на обучающей выборке: {test_loss}, Аккуратность: {test_acc}")
```

```python
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

```python
# summarize history for accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

```python
pred1 = model.predict(X_test)
pred1
```

```python
from sklearn.metrics import confusion_matrix
y_pred = np.argmax(pred1, axis=-1)
y_pred
print(y_pred)
y = np.argmax(Y_test, axis=1)
print(y)
```

```python
plt.figure(figsize=(15,8))
conf = confusion_matrix(y, y_pred)

ax=sns.heatmap(conf, annot=True, fmt='d')
ax.set_title('Матрица ошибок')
ax.set_xlabel('Предсказанные')
ax.set_ylabel('Фактические')
```

```python
from sklearn.metrics import precision_score, accuracy_score, recall_score, classification_report

print(precision_score(y, y_pred,  average='micro'))
print(accuracy_score(y, y_pred))
print(recall_score(y, y_pred,  average='micro'))
print(classification_report(y, y_pred))
```

```python
from keras.datasets import cifar10
(X_train,Y_train),(X_test,Y_test) = cifar10.load_data()
```

```python
print(X_train.shape)
Y_train=Y_train.flatten()
print(Y_train.shape)
print(X_test.shape)
Y_test=Y_test.flatten()
print(Y_test.shape)
```

```python
Y_train[0]
```

```python
fig, ax = plt.subplots(1,10, figsize=(15,8))

for i in range(10):
  img = X_train[Y_train==i][0]
  ax[i].imshow(img)
  ax[i].set_title(f"Цифра {i}")
```

```python
X_train = X_train / 255
X_test = X_test /255
```

```python

Y_train = keras.utils.to_categorical(Y_train, 10)
Y_test = keras.utils.to_categorical(Y_test, 10)

print(Y_train[0])
print(Y_test[0])
```

```python
from tensorflow.keras import layers
```

```python
import tensorflow as tf
import os
os.environ['CUDA_VISIBLE_DEVICES'] = '0'
tf.config.get_visible_devices()
```

```python
inputs = keras.Input(shape=(32, 32, 3), name="img")
x = layers.Conv2D(32, 3, activation="relu")(inputs)
x = layers.Conv2D(64, 3, activation="relu")(x)
block_1_output = layers.MaxPooling2D(3)(x)

x = layers.Conv2D(64, 3, activation="relu", padding="same")(block_1_output)
x = layers.Conv2D(64, 3, activation="relu", padding="same")(x)
block_2_output = layers.add([x, block_1_output])

x = layers.Conv2D(64, 3, activation="relu", padding="same")(block_2_output)
x = layers.Conv2D(64, 3, activation="relu", padding="same")(x)
block_3_output = layers.add([x, block_2_output])

x = layers.Conv2D(64, 3, activation="relu")(block_3_output)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Flatten()(x)
x = layers.Dense(256, activation="relu")(x)
x = layers.Dropout(0.5)(x)
outputs = layers.Dense(10, activation='softmax')(x)
```

```python
model3 = keras.Model(inputs, outputs, name="toy_resnet")
```

```python
model3.compile(optimizer='adam',
             loss='categorical_crossentropy',
             metrics=['accuracy'])

```

```python
model3.fit(X_train, Y_train, batch_size=64, epochs=15, validation_split=0.2)
```
